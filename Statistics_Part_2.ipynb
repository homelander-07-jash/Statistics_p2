# Statistics Part 2


1.  What is hypothesis testing in statistics ?

- Hypothesis testing is a method used to decide whether there is enough evidence to reject a claim (hypothesis) about a population parameter based on sample data.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis ?

- Null hypothesis (H‚ÇÄ): The default assumption that there is no effect or difference.

- Alternative hypothesis (H‚ÇÅ or Ha): A statement that contradicts the null and suggests an effect or difference exists.

3. What is the significance level in hypothesis testing, and why is it important ?

- The significance level (Œ±) is the probability of rejecting the null hypothesis when it is actually true (Type I error). Common values: 0.05 or 0.01. It sets the threshold for determining whether the P-value is "small enough" to reject H‚ÇÄ.

4. What does a P-value represent in hypothesis testing ?

- The P-value is the probability of obtaining a test statistic at least as extreme as the observed one, assuming the null hypothesis is true.

5. How do you interpret the P-value in hypothesis testing ?

- P ‚â§ Œ±: Reject the null hypothesis (statistically significant result).

- P > Œ±: Fail to reject the null hypothesis (not statistically significant).

6. What are Type 1 and Type 2 errors in hypothesis testing ?

- Type I Error (Œ±): Rejecting H‚ÇÄ when it is true.

- Type II Error (Œ≤): Failing to reject H‚ÇÄ when H‚ÇÅ is true.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing ?

- One-tailed test: Tests for an effect in one direction (e.g., greater than).

- Two-tailed test: Tests for an effect in either direction (e.g., not equal to).

8.  What is the Z-test, and when is it used in hypothesis testing ?

- The Z-test is used when:

- The population variance is known.

- The sample size is large (n ‚â• 30).
It tests whether the sample mean differs significantly from a known population mean.

10.  What is the T-distribution, and when should it be used instead of the normal distribution ?

- Use the T-distribution when:

- Sample size is small (n < 30).

- Population standard deviation is unknown.
It has heavier tails to account for increased variability in smaller samples.

12.  What is the T-test, and how is it used in hypothesis testing ?

- A T-test assesses whether the means of one or two groups are statistically different. Common types:

- One-sample T-test: Compare sample mean to population mean.

- Two-sample T-test: Compare means of two independent samples.

- Paired T-test: Compare means from the same group at different times.


13. What is the relationship between Z-test and T-test in hypothesis testing ?

- Both test hypotheses about means, but:

- T-test adjusts for small samples and unknown population SD.

- Z-test assumes a known population SD and large sample size.

14. What is a confidence interval, and how is it used to interpret statistical results ?

- A confidence interval (CI) provides a range of plausible values for a population parameter. A 95% CI means you can be 95% confident the interval contains the true parameter.

15. What is the margin of error, and how does it affect the confidence interval ?

- The margin of error is the range above and below the sample estimate. Larger margins mean wider confidence intervals, reflecting more uncertainty.

```python
# 1.  Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

from scipy.stats import norm
import numpy as np

# Parameters
population_mean = 100
population_std = 15
sample = np.array([102, 98, 95, 101, 104, 97, 96, 100, 99, 103])

sample_mean = np.mean(sample)
sample_size = len(sample)

# Z-test
z = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1 - norm.cdf(abs(z)))

print(f"Z-score: {z:.3f}")
print(f"P-value: {p_value:.3f}")

if p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

def visualize_type_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05):
    se = sigma / np.sqrt(n)
    x = np.linspace(85, 115, 500)

    null_dist = norm(mu0, se)
    alt_dist = norm(mu1, se)

    z_crit = norm.ppf(1 - alpha/2)
    x_crit_left = mu0 - z_crit * se
    x_crit_right = mu0 + z_crit * se

    plt.plot(x, null_dist.pdf(x), label="H‚ÇÄ Distribution", color="blue")
    plt.plot(x, alt_dist.pdf(x), label="H‚ÇÅ Distribution", color="green")

    plt.fill_between(x, null_dist.pdf(x), where=(x < x_crit_left) | (x > x_crit_right), color="red", alpha=0.3, label="Type I Error (Œ±)")
    plt.fill_between(x, alt_dist.pdf(x), where=(x_crit_left < x) & (x < x_crit_right), color="orange", alpha=0.3, label="Type II Error (Œ≤)")

    plt.axvline(x_crit_left, color='black', linestyle='--')
    plt.axvline(x_crit_right, color='black', linestyle='--')
    plt.title("Type I and II Errors")
    plt.legend()
    plt.grid()
    plt.show()

visualize_type_errors()
```

```python
# 9.  Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import scipy.stats as stats

def confidence_interval(sample, confidence=0.95):
    mean = np.mean(sample)
    se = stats.sem(sample)
    margin = se * stats.t.ppf((1 + confidence) / 2., len(sample) - 1)
    return mean - margin, mean + margin

ci = confidence_interval(sample)
print(f"95% Confidence Interval: {ci[0]:.2f} to {ci[1]:.2f}")
```

```python
# 10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

from scipy import stats
import numpy as np

def margin_of_error(data, confidence=0.95):
    n = len(data)
    std_err = stats.sem(data)
    z_score = stats.norm.ppf((1 + confidence) / 2)
    return z_score * std_err

sample = np.random.normal(100, 15, 50)
print(f"Margin of Error (95%): {margin_of_error(sample):.2f}")
```

```python
# 11.  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

# Bayes‚Äô Theorem: P(A|B) = [P(B|A) * P(A)] / P(B)
def bayes_theorem(prior_A, prob_B_given_A, prob_B):
    return (prob_B_given_A * prior_A) / prob_B

# Example
P_A = 0.01      # Disease
P_B_given_A = 0.99  # Positive test if diseased
P_B = 0.05      # Overall positive tests

posterior = bayes_theorem(P_A, P_B_given_A, P_B)
print(f"Posterior Probability: {posterior:.4f}")
```

```python
# 16.  Implement an F-test using Python to compare the variances of two random samples.

def f_test(sample1, sample2):
    var1, var2 = np.var(sample1, ddof=1), np.var(sample2, ddof=1)
    f = var1 / var2
    df1, df2 = len(sample1) - 1, len(sample2) - 1
    p = 1 - stats.f.cdf(f, df1, df2)
    return f, p

s1 = np.random.normal(0, 2, 30)
s2 = np.random.normal(0, 1, 30)
print("F-test:", f_test(s1, s2))
```

```python
# 17.  Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results.

from scipy.stats import f_oneway

g1 = np.random.normal(100, 10, 30)
g2 = np.random.normal(105, 10, 30)
g3 = np.random.normal(110, 10, 30)

f_stat, p = f_oneway(g1, g2, g3)
print(f"ANOVA F = {f_stat:.2f}, P = {p:.4f}")
```

```python
# 18. Perform a one-way ANOVA test using Python to compare the means of different groups and plot the results.

import seaborn as sns
import pandas as pd

data = pd.DataFrame({
    "value": np.concatenate([g1, g2, g3]),
    "group": ["G1"]*30 + ["G2"]*30 + ["G3"]*30
})

sns.boxplot(x="group", y="value", data=data)
plt.title("Group Mean Comparison via Boxplot")
plt.grid()
plt.show()
```

```python
# 19.  Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

from scipy.stats import shapiro, levene

def check_anova_assumptions(groups):
    for i, g in enumerate(groups, 1):
        stat, p = shapiro(g)
        print(f"Group {i} Normality: P = {p:.3f}")
    stat, p = levene(*groups)
    print(f"Equal Variance (Levene's Test): P = {p:.3f}")

check_anova_assumptions([g1, g2, g3])
```

```python
# 20.  Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.

import statsmodels.api as sm
from statsmodels.formula.api import ols

df = pd.DataFrame({
    "Score": np.random.randn(60),
    "Gender": ["M", "F"] * 30,
    "Method": ["A"]*20 + ["B"]*20 + ["C"]*20
})

model = ols('Score ~ C(Gender) + C(Method) + C(Gender):C(Method)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```
